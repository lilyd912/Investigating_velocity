```python
import numpy as np
import pandas as pd
from scipy.stats import linregress

valid = gdf["height"].notna() & gdf["dist_along_ctrline"].notna()

x = gdf.loc[valid, "dist_along_ctrline"]
y = gdf.loc[valid, "height"]

fit = linregress(x, y)

residual = y - (fit.intercept + fit.slope * x)

gdf_res = gdf.loc[valid].copy()
gdf_res["residual_m"] = residual
```

```python
gdf_res_ll = gdf_res.to_crs("EPSG:4326")
```

```python
#installing kml againnn cus i forgot it in the preamble
!pip install simplekml
import simplekml
```

```python
def residual_colour(value, max_abs=3):
    """
    Negative residuals = blue
    Near zero = white
    Positive residuals = red
    """

    value = max(-max_abs, min(max_abs, value))
    t = (value + max_abs) / (2 * max_abs)

    if t < 0.5:
        # blue to white
        frac = t / 0.5
        r = int(255 * frac)
        g = int(255 * frac)
        b = 255
    else:
        # white to red
        frac = (t - 0.5) / 0.5
        r = 255
        g = int(255 * (1 - frac))
        b = int(255 * (1 - frac))

    # KML format: alpha, blue, green, red
    return f"ff{b:02x}{g:02x}{r:02x}"
```

```python
kml = simplekml.Kml()

for idx, row in gdf_res_ll.iterrows():
    lon = row.geometry.x
    lat = row.geometry.y
    res = row["residual_m"]

    pnt = kml.newpoint(
        name=f"Point {idx}",
        coords=[(lon, lat)]
    )
#so this one is just using google earth points so we can indivudally see the values
#but its a bit difficult to use? and also makes the points look odd, but is useful for the values!
#instead i think a trend would be better here ?

    pnt.description = (
        f"Residual height: {res:.3f} m<br>"
        f"Height: {row['height']:.3f} m<br>"
        f"Distance along centreline: {row['dist_along_ctrline']}"
    )

    pnt.style.iconstyle.color = residual_colour(res, max_abs=3)
    pnt.style.iconstyle.scale = 0.7
```
